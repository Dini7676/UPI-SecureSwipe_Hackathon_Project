```python
import pandas as pd
from dataset_generator import generate
from feature_engineering import featurize
from lightgbm import LGBMClassifier
import joblib
df = generate(10000)
X = featurize(df)
y = df['is_fraud']
model = LGBMClassifier(n_estimators=200, max_depth=-1, learning_rate=0.05)
model.fit(X, y)
joblib.dump(model, 'model.pkl')
print('Saved model.pkl')
```
